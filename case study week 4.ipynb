# 3.Plot a bargraph of male vs female buyers in sales

```python
Female = cars_data['Buyer Gender'] =='Female'
no_of_females=Female.sum()
no_of_females
```

```python
Male = cars_data['Buyer Gender'] =='Male'
no_of_males=Male.sum()
no_of_males
```

# Bar Graph

```python
plt.bar(['Male','Female'],height = [no_of_males,no_of_females],color = ['Red','orange'],width = 0.5)
plt.grid()
plt.title('Male v/s Female buyers participated in sales')
plt.ylabel('No. of Participants')

```

# Insights

```python
When we plotted a bargraph of Males and Females buyers parcipated in Sales,we can see that here is a slight difference 
between the no of males and females.The no of females are more than males.
```
